```python
import cv2
import cvlib as cv
from google.colab.patches import cv2_imshow
```

```python
!wget -qq "https://kulapartners.b-cdn.net/wp-content/uploads/2017/06/multiple-personas-hero.jpg?x77431" -O 'input.jpg'
```

```python
image = cv2.imread("/content/input.jpg")
cv2_imshow(image)
```

```python
def detect_faces(image, threshold=0.5, enable_gpu=False):
    faces, confidences = cv.detect_face(image, threshold)

    for face in faces:
        image = cv2.rectangle(image, (face[0], face[1]), (face[2], face[3]), (0, 0, 255), 2)

    cv2_imshow(image)
    return faces, confidences
```

```python
image = cv2.imread("/content/input.jpg")
faces, confidences = detect_faces(image, threshold=0.45)
```
